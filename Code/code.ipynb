***Profession Specific Information Retrieval System***

**Import the necessary libraries**

```python
import os
import nltk
from nltk.stem import PorterStemmer
from nltk.tokenize import sent_tokenize, word_tokenize
from nltk.corpus import stopwords
```

**Defining list to store the corpus**

```python
files = []
```

```python
for i in os.listdir('Corpus'):
    if i.endswith('.txt'):
        files.append(i)
```

```python
print(files)
```

**Converting the corpus to lower case and tokenizing it. Removed tokens of length one from the list.** 

```python
f = open('Corpus\\'+files[1], 'r')
file_contents = f.read()
file_contents = word_tokenize(file_contents)
z = []
for x in file_contents:
    z.append(x.lower())
print(len(z))
for a in z:
    if(len(a) is 1):
        z.remove(a);
for a in z:
    if(len(a) is 1):
        z.remove(a);
```

```python
print(z)
f.close()
```

**Removing Stop Words from the list**

```python
sp_words = set(stopwords.words('english'))
sp_words.add('its')
for wrd in  z:
    if wrd in sp_words:
        z.remove(wrd)
```

```python
print(len(z))
print(z)
```

**Stemming and Lemmatization**

```python
pS = PorterStemmer() 
n = []
for w in z:
    n.append(pS.stem(w))
print(n)
```
